```python
def calcular_media(notas):
    if len(notas) == 0:
        return 0
    else:
        return sum(notas) / len(notas)

def main():
    notas = []
    
    while True:
        try:
            nota = float(input("Digite uma nota (ou um valor inválido para encerrar): "))
            
            if 0 <= nota <= 10:
                notas.append(nota)
            else:
                break  # Encerra o loop se a nota não estiver no intervalo válido
            
            print(f"Notas inseridas: {notas}")

            continuar = input("Deseja continuar inserindo notas? (S para Sim, qualquer outra tecla para encerrar): ")
            
            if continuar.upper() != 'S':
                break  # Encerra o loop se o usuário não quiser continuar
            
        except ValueError:
            print("Por favor, insira um valor numérico válido.")

    media = calcular_media(notas)
    
    if len(notas) > 0:
        print(f"\nMédia das notas: {media:.2f}")
    else:
        print("Nenhuma nota válida foi inserida.")

if __name__ == "__main__":
    main()
```

```python
def calcular_media(notas):
    if not notas:
        return 0
    return sum(notas) / len(notas)

def obter_notas():
    notas = []
    
    while True:
        try:
            nota = float(input("Digite uma nota (ou um valor inválido para encerrar): "))
            
            if 0 <= nota <= 10:
                notas.append(nota)
            else:
                break  
            
            print(f"Nota {len(notas)} adicionada: {nota}")
            
            continuar = input("Deseja adicionar outra nota? (S para Sim, qualquer outra tecla para encerrar): ")
            
            if continuar.upper() != 'S':
                break
            
        except ValueError:
            print("Por favor, insira um valor numérico válido.")

    return notas

def main():
    
    notas = obter_notas()
    
    if notas:
        media = calcular_media(notas)
        print(f"\nNotas inseridas: {notas}")
        print(f"Média das notas: {media:.2f}")
    else:
        print("Nenhuma nota válida foi inserida. O programa será encerrado.")

if __name__ == "__main__":
    main()
```
