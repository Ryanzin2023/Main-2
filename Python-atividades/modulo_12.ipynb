```python
#----------Inicio das funções---------------

def recebeNome () -> str:
    return input('Informe seu nome: ').upper()

def imprimeNomes(listaNomes: list):
    for nome in listaNomes:
        print(f'Nome {listaNomes.index(nome) + 1}: {nome}')

#----------Fim das funções---------------


#----------Inicio das Aplicações---------------
listaNomes = []

while True:
    nome = recebeNome()
    if recebeNome() != 'S':
        listaNomes.append(nome)
        print(f'Olá {nome}')
    else:
        print(f'Aplicação encerrada')
        break

imprimeNomes(listaNomes)

#----------Fim das Aplicações---------------
```
